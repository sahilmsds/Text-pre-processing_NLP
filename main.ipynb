```python
from nltk.tokenize import word_tokenize #Sentences into word tokens
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer #Stemming
from nltk.stem import WordNetLemmatizer #Lemmatization tool
from nltk.corpus import wordnet
from nltk import pos_tag
```

```python
input_text = """
Technology is rapidly! changing the world; it has transformed how we communicate,
work, and live. Artificial intelligence: a fascinating field, is being used in
healthcare, education, and business sectors... Have you ever wondered? The impact
is truly remarkable! Many researchers are working hard to develop new algorithms.

Nature, however, remains our greatest teacher; it has inspired countless inventions.
The forests, oceans, and mountains: all hold secrets waiting to be discovered!!
Scientists study these ecosystems carefully... but do we really understand them?
Every living creature plays an important role; none should be overlooked or ignored!
"""
```

```python
text_lower = input_text.lower()
tokens = word_tokenize(text=text_lower)
print(tokens)
```

Removing Punctuations first

```python
tokens_no_punc = [word for word in tokens if word.isalpha()]
print(f"Words before removing punctuation: {len(tokens)} | Words after removing punctuations: {len(tokens_no_punc)}")
print(tokens_no_punc)
```

Removing Stopwords

```python
stop_words = set(stopwords.words('english'))
tokens_no_stopwords = [word for word in tokens_no_punc if word not in stop_words]
print(f"Words before removing Stopwords: {len(tokens_no_punc)} | Words after removing Stopwords: {len(tokens_no_stopwords)}")
print(f"Total stopwords removed: {len(tokens_no_punc) - len(tokens_no_stopwords)}")
print(f'\n{tokens_no_stopwords}')
```

Stemming

```python
stemmer = PorterStemmer()
tokens_stemmed = [stemmer.stem(word) for word in tokens_no_stopwords]
print(f"TOkens Stemmed \n{tokens_stemmed}")
```

```python
print("\n    Word -> Stemmed Form (sample):")

for original, stemmed in zip(tokens_no_stopwords[:15], tokens_stemmed[:15]):
    if original != stemmed:
        print(f"   {original:<20} -->  {stemmed}")
```

Lemmatization

```python
lemmitizer = WordNetLemmatizer()
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
pos_tags = pos_tag(tokens_no_stopwords)
```

```python
pos_tags
```

```python
tokens_lemmatized = [lemmitizer.lemmatize(word, get_wordnet_pos(tag))
                     for word , tag in pos_tags]
print(f"Lemmatized Tokens\n{tokens_lemmatized}")
```

```python
print("     Word -> Lemmatized form (Sample)\n")
for original, lemmatized in zip(tokens_no_stopwords, tokens_lemmatized):
    if original != lemmatized:
        print(f"     {original:<20} --> {lemmatized}") 
```

## Assignment 2 

```python
preprocessed_text = ' '.join(tokens_lemmatized)
print(preprocessed_text)
```

Implementing Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform([preprocessed_text])
bow_array = bow.toarray()
features_names = cv.get_feature_names_out()
print(f"Features Names: {features_names}")
print(f"Bag of Words vector: {bow_array}") 
```

Implementing TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```python
tf_idf_matrix = tfidf.fit_transform([preprocessed_text])
tf_idf_array = tf_idf_matrix.toarray()
tf_features_names = tfidf.get_feature_names_out()
print(f"TF-IDF feature names: {tf_features_names}")
print(f"TF-IDF vector: {tf_idf_array}")
```

```python
# Below paragraph is taken from https://thefusepathway.com/blog/a-brief-history-of-ai/?gad_source=1&gad_campaignid=22311785187&gbraid=0AAAAApnnG7wfXOkpWIJmAruYK74cyBni2&gclid=Cj0KCQjw4PPNBhD8ARIsAMo-icxP52Dxx9WfYqiL6lfmIvU6Hw_rSSLuyZfCxDHglq8OvOn1x6g5GW0aAsXDEALw_wcB , The World’s Response to AI: Hot and Cold
new_para = """
In 1956, an AI workshop was held at Dartmouth College, and the field of artificial intelligence was officially founded as an academic discipline. Computers from the Dartmouth workshop were shown to be able to solve algebra word problems, learn English, play checkers, and more results that were stunning to the modern audiences of the time. Press coverage of the workshop created a media frenzy, and by the 1960s defense departments and labs all over the world had established research programs into artificial intelligence. Many scientists in the field believed that the creation of a full-blown artificially intelligent machine was right around the corner. Unfortunately, by the mid-1970s, many colossal technological constraints and limitations severely hindered progress. In 1974, the U.S. and British governments eliminated AI project funding due to budget considerations. This would later be known as the “AI winter.”
In the 1980s, computer technologies advanced significantly. One area of advancement was in “expert systems.” Expert systems are computer programs designed to mimic the thought process, logic, and decision-making of real human experts. Instead of conventional procedural code, expert systems employ “if-then” rules to reason through existing knowledge in order to solve problems. Implementing these expert systems was a widespread commercial success, reinvigorating the dying field of AI research and development. By the mid-80s, AI technology had become a billion dollar industry. Japanese government efforts to produce an $850 million dollar, “Epoch-making” AI supercomputer revived global government funding of AI research, but various failures (like the collapse of the Lisp Machine market) led to a second, longer, AI winter."""
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
sentences = sent_tokenize(new_para)
processed_sentences = []
for i in sentences:
    words = word_tokenize(i.lower())
    words = [w for w in words if w.isalpha()]
    words = [w for w in words if w not in stop_words]
    processed_sentences.append(words)
    
print(processed_sentences[:2])
```

Word2Vec 

```python
from gensim.models import Word2Vec
model = Word2Vec(processed_sentences, vector_size=100,window=8, min_count=1, workers=4, epochs=100)
```

```python
print(model.wv.key_to_index)
```

```python
similarity1 = model.wv.similarity('computer', 'systems')
similarity2 = model.wv.similarity('intelligence', 'artificial')
similarity3 = model.wv.similarity('ai', 'intelligence')

print(f"Similarity between words 'Computer' and 'Systems': {similarity1:.2f}")
print(f"Similarity between words 'Intelligence' and 'Artificial': {similarity2:.2f}")
print(f"Similarity between words 'ai' and 'intelligence: {similarity3:.2f}")
```
