```python
from nltk.tokenize import word_tokenize #Sentences into word tokens
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer #Stemming
from nltk.stem import WordNetLemmatizer #Lemmatization tool
from nltk.corpus import wordnet
from nltk import pos_tag
```

```python
input_text = """
Technology is rapidly! changing the world; it has transformed how we communicate,
work, and live. Artificial intelligence: a fascinating field, is being used in
healthcare, education, and business sectors... Have you ever wondered? The impact
is truly remarkable! Many researchers are working hard to develop new algorithms.

Nature, however, remains our greatest teacher; it has inspired countless inventions.
The forests, oceans, and mountains: all hold secrets waiting to be discovered!!
Scientists study these ecosystems carefully... but do we really understand them?
Every living creature plays an important role; none should be overlooked or ignored!
"""
```

```python
text_lower = input_text.lower()
tokens = word_tokenize(text=text_lower)
print(tokens)
```

Removing Punctuations first

```python
tokens_no_punc = [word for word in tokens if word.isalpha()]
print(f"Words before removing punctuation: {len(tokens)} | Words after removing punctuations: {len(tokens_no_punc)}")
print(tokens_no_punc)
```

Removing Stopwords

```python
stop_words = set(stopwords.words('english'))
tokens_no_stopwords = [word for word in tokens_no_punc if word not in stop_words]
print(f"Words before removing Stopwords: {len(tokens_no_punc)} | Words after removing Stopwords: {len(tokens_no_stopwords)}")
print(f"Total stopwords removed: {len(tokens_no_punc) - len(tokens_no_stopwords)}")
print(f'\n{tokens_no_stopwords}')
```

Stemming

```python
stemmer = PorterStemmer()
tokens_stemmed = [stemmer.stem(word) for word in tokens_no_stopwords]
print(f"TOkens Stemmed \n{tokens_stemmed}")
```

```python
print("\n    Word -> Stemmed Form (sample):")

for original, stemmed in zip(tokens_no_stopwords[:15], tokens_stemmed[:15]):
    if original != stemmed:
        print(f"   {original:<20} -->  {stemmed}")
```

Lemmatization

```python
lemmitizer = WordNetLemmatizer()
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
pos_tags = pos_tag(tokens_no_stopwords)
```

```python
pos_tags
```

```python
tokens_lemmatized = [lemmitizer.lemmatize(word, get_wordnet_pos(tag))
                     for word , tag in pos_tags]
print(f"Lemmatized Tokens\n{tokens_lemmatized}")
```

```python
print("     Word -> Lemmatized form (Sample)\n")
for original, lemmatized in zip(tokens_no_stopwords, tokens_lemmatized):
    if original != lemmatized:
        print(f"     {original:<20} --> {lemmatized}") 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
sentence_a = "Technology is rapidly changing the world"
sentence_b = "Artificial intelligence is used in healthcare education and business"
sentence_c = "Nature remains our greatest teacher and has inspired countless inventions"
 
three_sentences = [sentence_a, sentence_b, sentence_c]
```

```python
tfidf_sent = TfidfVectorizer()
tfidf_sent_matrix = tfidf_sent.fit_transform(three_sentences)
cos_sim_matrix = cosine_similarity(tfidf_sent_matrix)

print("COSINE SIMILARITY BETWEEN SENTENCES")
print(f"\nSentence 1: {sentence_a}")
print(f"Sentence 2: {sentence_b}")
print(f"Sentence 3: {sentence_c}\n")

print(f"Cosine Similarity (Sent1 vs Sent2): {cos_sim_matrix[0][1]:.4f}")
print(f"Cosine Similarity (Sent1 vs Sent3): {cos_sim_matrix[0][2]:.4f}")
print(f"Cosine Similarity (Sent2 vs Sent3): {cos_sim_matrix[1][2]:.4f}")
print(f"\nFull Similarity Matrix:\n{cos_sim_matrix}")
```

text classification

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
```

```python
texts = [
    "Artificial intelligence is transforming healthcare and education",
    "Machine learning algorithms help businesses make decisions",
    "Researchers are developing new deep learning models",
    "Technology is changing how we work and communicate",
    "Neural networks are used in image recognition systems",
    "Computers can now solve complex mathematical problems",
    "Software engineers build applications for mobile devices",
    "Data science uses statistics and programming together",
 
    "Forests and oceans hold many undiscovered secrets",
    "Scientists study ecosystems to understand living creatures",
    "Nature inspires inventions through its complex systems",
    "Mountains and rivers shape the geography of our planet",
    "Every living creature plays an important role in nature",
    "Plants and animals adapt to survive in their environments",
    "Biodiversity is essential for a healthy ecosystem",
    "Researchers study birds, insects, and marine life carefully",
]
```

```python
labels = [
    "Technology", "Technology", "Technology", "Technology",
    "Technology", "Technology", "Technology", "Technology",
    "Nature",     "Nature",     "Nature",     "Nature",
    "Nature",     "Nature",     "Nature",     "Nature",
]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    texts, labels, test_size=0.25, random_state=42
)
```

```python
nb_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf',   MultinomialNB())
])
 
nb_pipeline.fit(X_train, y_train)
y_pred = nb_pipeline.predict(X_test)
```

```python
print(f"\nTraining samples: {len(X_train)} | Test samples: {len(X_test)}")
print(f"\nAccuracy: {accuracy_score(y_test, y_pred):.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
new_sentences = [
    "Deep learning is revolutionizing computer vision",
    "The ocean ecosystem is full of undiscovered life",
]
new_predictions = nb_pipeline.predict(new_sentences)
print("Predictions on new sentences:")
for sent, pred in zip(new_sentences, new_predictions):
    print(f"  '{sent}'\n   --> Predicted Category: {pred}\n")
```

Text Clustering

```python
from sklearn.cluster import KMeans
import numpy as np
tfidf_cluster = TfidfVectorizer()
X_cluster = tfidf_cluster.fit_transform(texts)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(X_cluster)
cluster_labels = kmeans.labels_
```

```python
print(f"\nCluster assignments for each document:")
for i, (text, cluster) in enumerate(zip(texts, cluster_labels)):
    print(f"  Cluster {cluster} | {text[:60]}...")
```

```python
print("\nTop keywords per cluster:")
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
terms = tfidf_cluster.get_feature_names_out()

for cluster_id in range(2):
    top_terms = [terms[ind] for ind in order_centroids[cluster_id, :8]]
    print(f"  Cluster {cluster_id}: {', '.join(top_terms)}")
```

```python
print("\nCluster vs Actual Label comparison:")
for text, cluster, actual in zip(texts, cluster_labels, labels):
    print(f"  Cluster {cluster} | Actual: {actual:<12} | {text[:50]}...")
```
