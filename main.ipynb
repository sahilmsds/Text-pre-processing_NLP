```python
from nltk.tokenize import word_tokenize #Sentences into word tokens
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer #Stemming
from nltk.stem import WordNetLemmatizer #Lemmatization tool
from nltk.corpus import wordnet
from nltk import pos_tag
```

```python
input_text = """
Technology is rapidly! changing the world; it has transformed how we communicate,
work, and live. Artificial intelligence: a fascinating field, is being used in
healthcare, education, and business sectors... Have you ever wondered? The impact
is truly remarkable! Many researchers are working hard to develop new algorithms.

Nature, however, remains our greatest teacher; it has inspired countless inventions.
The forests, oceans, and mountains: all hold secrets waiting to be discovered!!
Scientists study these ecosystems carefully... but do we really understand them?
Every living creature plays an important role; none should be overlooked or ignored!
"""
```

```python
text_lower = input_text.lower()
tokens = word_tokenize(text=text_lower)
print(tokens)
```

Removing Punctuations first

```python
tokens_no_punc = [word for word in tokens if word.isalpha()]
print(f"Words before removing punctuation: {len(tokens)} | Words after removing punctuations: {len(tokens_no_punc)}")
print(tokens_no_punc)
```

Removing Stopwords

```python
stop_words = set(stopwords.words('english'))
tokens_no_stopwords = [word for word in tokens_no_punc if word not in stop_words]
print(f"Words before removing Stopwords: {len(tokens_no_punc)} | Words after removing Stopwords: {len(tokens_no_stopwords)}")
print(f"Total stopwords removed: {len(tokens_no_punc) - len(tokens_no_stopwords)}")
print(f'\n{tokens_no_stopwords}')
```

Stemming

```python
stemmer = PorterStemmer()
tokens_stemmed = [stemmer.stem(word) for word in tokens_no_stopwords]
print(f"TOkens Stemmed \n{tokens_stemmed}")
```

```python
print("\n    Word -> Stemmed Form (sample):")

for original, stemmed in zip(tokens_no_stopwords[:15], tokens_stemmed[:15]):
    if original != stemmed:
        print(f"   {original:<20} -->  {stemmed}")
```

Lemmatization

```python
lemmitizer = WordNetLemmatizer()
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
pos_tags = pos_tag(tokens_no_stopwords)
```

```python
pos_tags
```

```python
tokens_lemmatized = [lemmitizer.lemmatize(word, get_wordnet_pos(tag))
                     for word , tag in pos_tags]
print(f"Lemmatized Tokens\n{tokens_lemmatized}")
```

```python
print("     Word -> Lemmatized form (Sample)\n")
for original, lemmatized in zip(tokens_no_stopwords, tokens_lemmatized):
    if original != lemmatized:
        print(f"     {original:<20} --> {lemmatized}") 
```
